# VARIABLE ANALYIS - UNIVARIATE, BIVARIATE, MULTIVARIATE

```python
numerical = ["Age","RestingBP","Cholestrol","MaxHeartRateAchieve","STDepression"]
numerical_axis = ["Age of Patient","Resting Blood Pressure", "Cholestrol", "Maximum Heart Rate Achived","ST Depression"]
```

```python
list(zip(numerical,numerical_axis))
```

```python
title_font = {"family": "arial", "color":"darkblue","weight":"bold","size":15}
axis_font = {"family": "arial", "color":"darkred","weight":"bold","size":10}
for i,z in list(zip(numerical,numerical_axis)):
    plt.figure(figsize = (8,8), dpi = 90)
    sns.distplot(df[i], hist_kws=dict(linewidth=1, edgecolor = 'r'), bins = 30)
    
    plt.title(i, fontdict = title_font)
    plt.xlabel(z, fontdict= axis_font)
    plt.ylabel("Density", fontdict = axis_font)
    
    plt.tight_layout()
    plt.show()
```

# CATEGORICAL VARIABLE(ANALYSIS WITH PIE CHART)

```python
categorical = ['Gender', 'ChestPain', 'RestingBP', 'Rest_ECG', 'ExcerciseIndusedAngina', 'slope', 'NoofMajorVessels', 'thal', 'target']
```

```python
categorical_axis_name = ["Gender","Chest Pain Type","Fasting Blood Sugar", "Resting ElectroCardiographic Results",
                        "Excerise Induced Angina", "The Slope of ST Segment", "Number of Major Vessels","Thal","Target"]
```

```python
title_font = {"family": "arial", "color":"darkblue","weight":"bold","size":15}
axis_font = {"family": "arial", "color":"darkred","weight":"bold","size":10}
for i,z in list(zip(categorical,categorical_axis_name)):
    fig, axis = plt.subplots(figsize=(8,6))
    
    observational_values = list(df[i].value_counts().index)
    toatl_observation_values = list(df[i].value_counts())
    
    axis.pie(toatl_observation_values,labels= observational_values, autopct = "%1.1f%%",startangle = 110 , labeldistance = 1.1)
    axis.axis("equal")
    
    plt.title((i+ "(" + z + ")"), fontdict = title_font)
    plt.legend()
    plt.show()
```

# BIVARIATE ANALYSIS

```python
numerical
```

```python
numerical.append("target") # Analyze target with all numerical variables
numerical 
```

```python
title_font = {"family": "arial", "color":"darkgreen","weight":"bold","size":15}
axis_font = {"family": "arial", "color":"darkred","weight":"bold","size":10}
for i,z in list(zip(numerical,numerical_axis)):
    graph = sns.FacetGrid(df[numerical], hue = "target", height =7, xlim = (df[i].min()-10, (df[i].max() + 10)))
    graph.map(sns.kdeplot, i, shade = True)
    graph.add_legend()
    
    
    plt.title(i, fontdict = title_font)
    plt.xlabel(z, fontdict= axis_font)
    plt.ylabel("Density", fontdict = axis_font)
    
    plt.tight_layout()
    plt.show()
```

# Categorical Variables - Target Variable(Analysis with Count Plot and Violinplot)


```python
categorical
```

```python
title_font = {"family": "arial", "color":"darkgreen","weight":"bold","size":15}
axis_font = {"family": "arial", "color":"darkred","weight":"bold","size":10}
for i,z in list(zip(categorical,categorical_axis_name)):
    plt.figure(figsize = (8,5))
    sns.countplot(i,data = df[categorical], hue = "target")
    
    plt.title(i, fontdict = title_font)
    plt.xlabel(z, fontdict= axis_font)
    plt.ylabel("Density", fontdict = axis_font)
    
    plt.tight_layout()
    plt.show()
```

```python
title_font = {"family": "arial", "color":"darkgreen","weight":"bold","size":15}
axis_font = {"family": "arial", "color":"darkred","weight":"bold","size":10}
for i,z in list(zip(categorical,categorical_axis_name)):
    plt.figure(figsize = (8,5))
    sns.violinplot(i,data = df[categorical], hue = "target")
    sns.color_palette(palette="copper")
    
    plt.title(i, fontdict = title_font)
    plt.xlabel(z, fontdict= axis_font)
    plt.ylabel("Density", fontdict = axis_font)
    
    plt.tight_layout()
    plt.show()
```
